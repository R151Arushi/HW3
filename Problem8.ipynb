```python
import numpy as np
```

```python
mnist = np.load('/Users/arushisharma/Desktop/hw3-2023/data/mnist-data-hw3.npz')
train_data, train_labels = mnist['training_data'], mnist['training_labels']
test_data, test_labels = mnist['test_data'], mnist['training_labels']

```

```python
# Reshape train_data to 2D array
num_samples, num_channels, num_rows, num_cols = train_data.shape
train_data_2d = train_data.reshape(num_samples, num_channels*num_rows*num_cols)

# Compute mean and covariance for each digit class
num_classes = len(np.unique(train_labels))
means = []
covs = []
for i in range(num_classes):
    class_data = train_data_2d[train_labels == i]
    class_mean = np.mean(class_data, axis=0)
    class_cov = np.cov(class_data.T)
    means.append(class_mean)
    covs.append(class_cov)
```

```python
import matplotlib.pyplot as plt

# Compute mean and covariance for each digit class
num_classes = len(np.unique(train_labels))
means = []
covs = []
for i in range(num_classes):
    class_data = train_data_2d[train_labels == i]
    class_mean = np.mean(class_data, axis=0)
    class_cov = np.cov(class_data.T)
    means.append(class_mean)
    covs.append(class_cov)

digits = np.unique(train_labels)
# Visualize covariance matrix for digit 0
for d in digits:
    cov = covs[d]
    plt.imshow(cov, cmap='viridis')
    plt.colorbar()
    plt.title(f'Covariance matrix for digit {d}')
    plt.show()
```
